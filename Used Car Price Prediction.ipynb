```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_csv('Data_Train.csv')
```

```python
p,q=df.shape
print(p,q)
```

```python
#Removing NaN values
df=df.dropna()
p,q=df.shape
print(p,q)
```

```python
df.index = range(df.shape[0])
```

```python
#Preprocessing
locval=set()
for i in range(df.shape[0]):
    locval.add(df['Location'][i])
```

```python
locval
```

```python
for i in locval:
    df[i]=pd.Series(0, index=df.index)
```

```python
for i in range(df.shape[0]):
    a=df['Location'][i]
    df[a][i]=1
```

```python
df.head(5)
```

```python
del df['Location']
df.shape
```

```python
for i in range(df.shape[0]):
    a=str(df['Year'][i])
    df['Year'][i]=int(a[2:])
```

```python
df.head(5)
```

```python
fuelval=set()
for i in range(df.shape[0]):
    fuelval.add(df['Fuel_Type'][i])
print(fuelval)
for i in fuelval:
    df[i]=pd.Series(0, index=df.index)
for i in range(df.shape[0]):
    a=df['Fuel_Type'][i]
    df[a][i]=1
```

```python
del df['Fuel_Type']
df.shape
```

```python
transval=set()
for i in range(df.shape[0]):
    transval.add(df['Transmission'][i])
print(transval)
for i in transval:
    df[i]=pd.Series(0, index=df.index)
for i in range(df.shape[0]):
    a=df['Transmission'][i]
    df[a][i]=1
```

```python
del df['Transmission']
df.shape
```

```python
df['Owner_Type'].replace({"First":3, "Second":2, "Third":1}, inplace=True)
```

```python
df.head(10)
```

```python
for i in range(df.shape[0]):
    a=str(df['Mileage'][i])
    b=a.find('km')
    df['Mileage'][i]=a[:b-1]
```

```python
df['Engine'] = df['Engine'].map(lambda x: x.rstrip('CC'))
df['Power'] = df['Power'].map(lambda x: x.rstrip('bhp'))
```

```python
df['New_Price'] = df['New_Price'].map(lambda x: x.rstrip('Lakh'))
```

```python
df.head(10)
```

```python
from sklearn import preprocessing
x = df[['Kilometers_Driven']].values.astype(float)

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(x)
aba=np.ravel(x_scaled)
df['KM']=pd.Series(aba, index=df.index)
```

```python
del df['Kilometers_Driven']
```

```python
x = df[['Mileage']].values.astype(float)

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(x)
aba=np.ravel(x_scaled)
df['Mil']=pd.Series(aba, index=df.index)
x = df[['Engine']].values.astype(float)

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(x)
aba=np.ravel(x_scaled)
df['Eng']=pd.Series(aba, index=df.index)
x = df[['Power']].values.astype(float)

# Create a minimum and maximum processor object
min_max_scaler = preprocessing.MinMaxScaler()

# Create an object to transform the data to fit minmax processor
x_scaled = min_max_scaler.fit_transform(x)
aba=np.ravel(x_scaled)
df['Pow']=pd.Series(aba, index=df.index)
```

```python
del df['Mileage']
del df['Engine']
del df['Power']
```

```python
df.head(10)
```

```python
brand=set()
for i in range(df.shape[0]):
    a=df['Name'][i]
    b=a.find(' ')
    brand.add(a[0:b])
brand
```

```python
for i in brand:
    df[i]=pd.Series(0, index=df.index)
for i in range(df.shape[0]):
    a=df['Name'][i]
    b=a.find(' ')
    df[a[0:b]][i]=1
    
```

```python
df.head(10)
```

```python
df=df[[c for c in df if c not in ['New_Price','Price']] 
       + ['New_Price', 'Price']]
```

```python
del df['Name']
```

```python
#Using Linear Regression with Least Square Method
x=df.iloc[0:,:48].values
y=df.iloc[0:,48:].values
print(x[0])
#Using Linear Regression
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 2/10, random_state = 0)
lin=LinearRegression()
lin.fit(x_train,y_train)
```

```python
for i in range(df.shape[0]):
    a=df['New_Price'][i]
    b=a.find('Cr')
    if(b!=-1):
        df['New_Price'][i]=a[:b-1]
```

```python
for i in range(df.shape[0]):
    a=df['New_Price'][i]
    b=a.find(' ')
    if(b!=-1):
       df['New_Price'][i]=a[:b] 
```

```python
df.head(20)
```

```python
y_pred=lin.predict(x_test)
lin.score(x_train,y_train)
lin.score(x_test,y_test)
```

```python
y_pred=lin.predict(x_test)
db1 = pd.DataFrame({'Actual': np.ravel(y_test), 'Predicted': np.ravel(y_pred)})
db2 = db1.head(25)
db2
```

```python
#Using Decision Tree Regressor
from sklearn.tree import DecisionTreeRegressor  
  
# create a regressor object 
regressor = DecisionTreeRegressor(random_state = 0)  
  
# fit the regressor with X and Y data 
regressor.fit(x_train, y_train)
```

```python
y_pred=regressor.predict(x_test)
regressor.score(x_train,y_train)
regressor.score(x_test,y_test)
```

```python
y_pred=regressor.predict(x_test)
db1 = pd.DataFrame({'Actual': np.ravel(y_test), 'Predicted': np.ravel(y_pred)})
db2 = db1.head(25)
db2
```

```python
#Using Random Forest Regressor
from sklearn.ensemble import RandomForestRegressor 
  
 # create regressor object 
regressor1 = RandomForestRegressor(n_estimators = 100, random_state = 0) 
  
# fit the regressor with x and y data 
regressor1.fit(x_train, y_train)
print(regressor1.score(x_train, y_train))
print(regressor1.score(x_test, y_test))
```

```python
y_pred=regressor1.predict(x_test)
db1 = pd.DataFrame({'Actual': np.ravel(y_test), 'Predicted': np.ravel(y_pred)})
db2 = db1.head(25)
db2
```
